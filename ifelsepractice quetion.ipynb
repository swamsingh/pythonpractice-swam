```python
# 2.	Write a program that will determine weather when the value of temperature and humidity is provided by the user.
temperature = float(input("Enter the temperature (in °C): "))
humidity = float(input("Enter the humidity (%): "))

if temperature > 30 and humidity > 70:
    print("It's hot and humid.")
elif temperature > 30:
    print("It's hot.")
elif humidity > 70:
    print("It's humid.")
else:
    print("The weather is pleasant.")
```

```python
# 3.	Write a program that will take user input of cost price and selling price and determines whether its a loss or a profit
cost_price = float(input("Enter the cost price: "))
selling_price = float(input("Enter the selling price: "))

if selling_price > cost_price:
    print("Profit")
elif selling_price < cost_price:
    print("Loss")
else:
    print("No Profit No Loss")
```

```python
# 4.	Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.
heads = int(input("Enter the total number of heads: "))
legs = int(input("Enter the total number of legs: "))

chickens = (4 * heads - legs) // 2
dogs = heads - chickens

if chickens >= 0 and dogs >= 0:
    print(f"Chickens: {chickens}, Dogs: {dogs}")
else:
    print("Invalid input.")
```

```python
# 5.	Program to check whether a number is divisible by 5 and 11 or not
number = int(input("Enter a number: "))
if number % 5 == 0 and number % 11 == 0:
    print("The number is divisible by 5 and 11.")
else:
    print("The number is not divisible by 5 and 11.")
```

```python
# 6.	Write a program to check whether triangle is valid or not if angles are given.
angles = [float(input(f"Enter angle {i+1}: ")) for i in range(3)]
if sum(angles) == 180 and all(angle > 0 for angle in angles):
    print("The triangle is valid.")
else:
    print("The triangle is not valid.")
```

```python
# 7.	Write a program to check whether triangle is equilateral, scalene or isosceles
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))

if a == b == c:
    print("The triangle is equilateral.")
elif a == b or a == c or b == c:
    print("The triangle is isosceles.")
else:
    print("The triangle is scalene.")
```

```python
# 8.	Write a program to enter basic salary and calculate gross salary of an employee a.
basic_salary = float(input("Enter the basic salary: "))
hra = 0.1 * basic_salary
da = 0.2 * basic_salary
gross_salary = basic_salary + hra + da
print(f"Gross Salary: {gross_salary}")
```

```python
# 9.	Write a program to calculate the total electricity bill based on the following conditions:
units = float(input("Enter the number of units: "))

if units <= 50:
    bill = units * 2
elif units <= 150:
    bill = 50 * 2 + (units - 50) * 5
elif units <= 300:
    bill = 50 * 2 + 100 * 5 + (units - 150) * 10
else:
    bill = 50 * 2 + 100 * 5 + 150 * 10 + (units - 300) * 15

print(f"Total electricity bill: ₹{bill}")
```

```python
# 10.	Write a program to check whether any year is a leap year or not
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```
