```python
#Q3 Write a program to check if a given year is a leap year. A leap year is divisible by 4 but not by 100
#unless it is also divisible by 400.
year = int(input("enter year here : "))
if (year %4 == 0 and year% 100 != 0) or (year% 100 == 0 and year %400 == 0):
  print("leap year ")
else:
  print(" not a leap year")
```

```python
# Q6 Write a basic calculator program that takes two numbers and an operator (+, -, *, /) as input and
# performs the specified operation. Print the result based on the operation.
x = input("enter your number")
y = input("enter yout number")
ask  = input("enter your operation")
if ask == "+":
  print(int(x) + int(y))
elif ask == "-":
  print(int(x) - int(y))
elif ask == "*":
  print(int(x) * int(y))
elif ask == "/":
  print(int(x) / int(y))
elif ask == "%":
  print(int(x) % int(y))
else:
  print("invalid input")

```

```python
# Q10 Write a program that calculates the Body Mass Index (BMI) based on user input for weight (in
# kilograms) and height (in meters). Then categorize the BMI into:
# Underweight (BMI < 18.5)
# Normal weight (18.5 <= BMI < 24.9)
# Overweight (25 <= BMI < 29.9)
# Obesity (BMI >= 30)
# Use the formula: BMI = weight / (height ** 2)
weight = float(input("enter your weight in KG here : "))
height = float(input("enter youy height in Meters here : "))

BMI = (weight/(height**2))
if (BMI <18.5 ):
  print(f"your BMI is {BMI} you are inderweight")
elif(18.5 <= BMI < 24.9):
  print(f"your BMI is {BMI}you have normal weight")
elif(25 <= BMI <= 29.9):
  print(f"your bmi is {BMI}you are overweight")
else:
  print(f" your BMI is {BMI}you have obesity")




```
