```python
# q.1
a = [34, 1, 78, 45, 900, 3]
largest = a[0]  # Initialize to the first element of the list
for i in range(len(a)):
    if a[i] > largest:
        largest = a[i]
print("Largest number:", largest)
```

```python
a = []
size = int(input("enter size: "))
for i in range(size):
  val = input("enter item:")
  a.append(val)
print(a)
max = a[0]
for i in range(len(a)):
  if(a[i]>max):
      max = a[i]
print("largest number:",max)
```
