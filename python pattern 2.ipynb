```python
# print a right angle triangle pattern
for row in range(1,6):
  for col in range(1,row+1):
    print("*",end=" ")
  print()
```

```python
n = 6
for row in range (1,n+1):
  num= 65
  for x in range(row,n+1):
    print(chr(num),end=" ")
  print()
```

```python
n = 6
for row in range (1,n+1):
  num= 65
  for x in range(row,n+1):
    print(chr(num),end=" ")
    num= num+1
  print()
```

```python
num = 20
for i in range(4):
  for j in range(4-i):
    print(num, end=" ")
    num = num+1
  print()
```
